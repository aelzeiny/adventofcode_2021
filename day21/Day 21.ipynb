```python
die_roll = 0
def deterministic_100():
    global die_roll
    i = 0
    while True:
        die_roll += 1
        yield (i % 100) + 1
        i += 1
```

```python
class Player:
    def __init__(self, name: str, starting: int):
        self.pos = starting
        self.score = 0
        self.name = name
        
    def step(self, steps):
        self.pos = (self.pos + steps) % 10
        self.score = (self.pos + 1) + self.score
        print(f'{self.name} stepped {steps}. Score: {self.score}')
```

```python
ps = [Player('p1', 4), Player('p2', 9)]
p_i = 0
die_roll = 0
die = deterministic_100()
while all(p.score < 1000 for p in ps):
    die_rolls = [next(die) for _ in range(3)]
    print(die_rolls)
    p_draw = sum(die_rolls)
    p = ps[p_i]
    p.step(p_draw)
    p_i = (p_i + 1) % len(ps)
```

```python
print(die_roll * ps[p_i].score)
```
