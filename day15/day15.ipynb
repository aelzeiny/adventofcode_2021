```python
with open('./day15_input.txt') as f:
    grid = [[int(c) for c in l.strip()] for l in f.readlines()]
```

```python
def expand_one(num):
    return [[((num + c + r) % 9 if (num + c + r) % 9 != 0 else 9) for c in range(5)] for r in range(5)]

def expand_kong(grid):
    answer = [[0 for _ in range(len(grid[r % len(grid)]) * 5)] for r in range(len(grid) * 5)]
    for r in range(len(grid)):
        for c in range(len(grid[r])):
            expanded = expand_one(grid[r][c])
            for rr in range(len(expanded)):
                for cc in range(len(expanded[rr])):
                    answer[len(grid) * rr + r][len(grid[r]) * cc + c] = expanded[rr][cc]
    return answer


expand_one(8)
```

```python
from dijkstar import Graph, find_path

def assess_risk(grid):
    def rc_to_key(r, c):
        return r * len(grid) + c

    def key_to_rc(n):
        return n // len(grid), n % len(grid)


    g = Graph()
    for r in range(len(grid)):
        for c in range(len(grid[r])):
            lr = r - 1
            lc = c - 1
            ur = r + 1
            uc = c + 1
            if lr >= 0:
                g.add_edge(rc_to_key(r, c), rc_to_key(lr, c), grid[r][c])
            if lc >= 0:
                g.add_edge(rc_to_key(r, c), rc_to_key(r, lc), grid[r][c])
            if ur < len(grid):
                g.add_edge(rc_to_key(r, c), rc_to_key(ur, c), grid[r][c])
            if uc < len(grid[r]):
                g.add_edge(rc_to_key(r, c), rc_to_key(r, uc), grid[r][c])
    path_info = find_path(g, rc_to_key(0, 0), rc_to_key(len(grid) - 1, len(grid[0]) - 1))
    risk = 0
    for n in path_info.nodes[1:]:
        r, c = key_to_rc(n)
        print(r, c)
        risk += grid[r][c]
    return risk
```
